This notebook contains shaking observations from the Wichmann Catalog in order of their appearance. MMI explainer available at: https://earthquake.usgs.gov/learn/topics/mercalli.php

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
from shake_gauge import shake_gauge
from pandas import to_pickle, read_pickle
```

```python
%matplotlib inline
MMI = np.linspace(0, 12, 1000)
shake_gauges = list()
```

### 1. Labuha, Batjan Island

"4:30, 7, 10, 10:30 a.m. and 6:30 p.m. Labuha, island of Batjan. Each time a strong shock; the most powerful among them occurred on the days from the 25th-28th. The flag pole and the trees swayed to and fro. A flood wave was not seen."

#### MMI Estimate:
Min: 4, Mean: 4.5, Max: 5

```python
name = 20001
city_name = 'Labuha'
lat = -0.633516
lon = 127.485009
VS30 = 255.53
kind = 'norm'
params = [4.5, .7]

labuha = shake_gauge(name, lon, lat, VS30, kind, params, city_name)
shake_gauges.append(labuha)

plt.plot(MMI, labuha.distribution.pdf(MMI))
plt.xlabel('MMI')
plt.ylabel('Probability Density')
plt.title('Labuha')
plt.show()
```
